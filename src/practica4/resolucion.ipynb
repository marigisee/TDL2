Para entender la diferencia en los tiempos de ejecución de cada algoritmo de concatenación hay que entender la naturaleza de la clase String: **es inmutable**. Por lo tanto, al ser inmutable, si queremos realizar una concatenación sobre una variable String definida en nuestro programa, Java **crea nuevas instancias** de la variable adicionando los valores que queremos concatenar.

Por lo tanto, las clases ``StringBuffer`` y ``StringBuilder`` son **clases mutables**, por lo que si queremos modificar el valor de una instancia de las mismas no es necesario crear una nueva instancia sino que modifica la existente.

La diferencia entre ``StringBuffer`` y ``StringBuilder`` es la seguridad en cuanto a hilos (*Thread-Safety*): si nos encontramos desarrollando una aplicación que va aa ejecutarse en un solo hilo será mejor usar ``StringBuilder`` debido a que es más eficiente.
